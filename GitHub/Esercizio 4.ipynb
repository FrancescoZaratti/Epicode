```python
print ('Programma per calcolare media, massimo e minimo, riguardo la pressione sanguigna diastolica, per anno di età (0-81)\n')

age = int (input ('Selezione età :'))

mask = dbt ['Age (years)'] == age #?? Operatore per fare il compreso 10>x<30

mask_def = dbt [mask] ['Diastolic blood pressure (mm Hg)'].mean ()

if pd.isna(mask_def) == 0:
    print ('\nValore medio della pressione sanguigna diastolica per la fascia di età selezionata :',mask_def,'mm Hg')
else:
    print ('\nIl valore media per la fascia di età selezionata non è disponibile')
```

```python
print ('Programma per calcolare media, massimo e minimo, riguardo la pressione sanguigna diastolica, per fasce di età (0-81)\n')

age_init = int (input ('Selezione età minima :'))
age_end = int (input ('Selezione età massima :'))

mask_1 = dbt ['Age (years)'] >= age_init
mask_2 = dbt ['Age (years)'] <= age_end

mask_mean = dbt [(mask_1) & (mask_2)] ['Diastolic blood pressure (mm Hg)'].mean ()
mask_max = dbt [(mask_1) & (mask_2)] ['Diastolic blood pressure (mm Hg)'].max ()
mask_min = dbt [(mask_1) & (mask_2)] ['Diastolic blood pressure (mm Hg)'].min ()

if pd.isna(mask_mean) == 0:
    print ('\nValore medio della pressione sanguigna diastolica per la fascia di età selezionata :',mask_mean,'mm Hg')
else:
    print ('\nIl valore media per la fascia di età selezionata non è disponibile')

if pd.isna(mask_min) == 0:
    print ('Valore massimo della pressione sanguigna diastolica per la fascia di età selezionata :',mask_min,'mm Hg')
else:
    print ('Il valore massimo per la fascia di età selezionata non è disponibile')

if pd.isna(mask_max) == 0:
    print ('Valore minimo della pressione sanguigna diastolica per la fascia di età selezionata :',mask_max,'mm Hg')
else:
    print ('Il valore minimo per la fascia di età selezionata non è disponibile')
```
